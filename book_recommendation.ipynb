```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
import scipy
```

```python
books = pd.read_csv('book.csv')
```

```python
books.head()
```

```python
df = books.iloc[:,1:]
```

```python
df.head()
```

```python
df=df.rename({'User.ID':'ID','Book.Title':'Title','Book.Rating':'Rating'},axis=1)
```

```python
df.head()
```

```python
df.sort_values(['ID'])
```

```python
df.shape
```

```python
len(df['ID'].unique())   # no of unique user ID
```

```python
len(df['Title'].unique())  # no of unique books
```

```python
# long data to wide data conversion using pivot table
```

```python
df2 = df.pivot_table(index='ID',columns='Title',values='Rating').reset_index(drop=True)
```

```python
df2
```

```python
# replacing index values by unique user ID
```

```python
df2.index = df['ID'].unique()
```

```python
df2
```

```python
# Impute those NaNs with 0 values
```

```python
df2.fillna(0,inplace=True)
df2
```

```python
# calculating Cosine similarity between users on array data
```

```python
user_sim = 1-pairwise_distances(df2.values,metric = 'cosine')
user_sim
```

```python
# store the results in a dataframe format
```

```python
user_sim2 = pd.DataFrame(user_sim)
user_sim2
```

```python
# set the index and column names to user ids
```

```python
user_sim2.index = df['ID'].unique()
user_sim2.columns = df['ID'].unique()
user_sim2
```

```python
# making diagonal values null
```

```python
np.fill_diagonal(user_sim,0)
user_sim2
```

```python
#  most similar users
```

```python
user_sim2.idxmax(axis=1)
```
